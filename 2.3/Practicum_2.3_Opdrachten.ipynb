Geef een overzicht met daarin het aantal producten waarvan de marge lager is dan de gemiddelde marge van alle producten samen. Geef in het overzicht tevens aan wat de gemiddelde marge is van dit aantal producten waarvan de marge lager dan de gemiddelde marge van alle producten samen is. (1 kolom, 2 rijen) 

```python
newDf = product.copy()
counter = 0
product['MARGIN'] = pd.to_numeric(product['MARGIN'], errors='coerce')
gemMargin = product['MARGIN'].mean()

for index, row in product.iterrows():
  marge = product.at[index, 'MARGIN']
  if marge < gemMargin:
    newDf.at[index, 'lagerOfHogerDanGemiddeld'] = 'lager'
    counter += 1
  elif marge > gemMargin:
    newDf.at[index, 'lagerOfHogerDanGemiddeld'] = 'hoger'

newDf['GEM_VAN_LAGER'] = newDf[newDf['lagerOfHogerDanGemiddeld'] == 'lager']['MARGIN'].mean()
newDf['AANTAL_LAGER'] = counter
newDf.loc[:, ['AANTAL_LAGER', 'GEM_VAN_LAGER']].drop_duplicates()
```

```python
newSalesStaff = sales_staff.copy()

for index, row in sales_staff.iterrows():
    if 'Manager' in str(row['POSITION_EN']):
        newSalesStaff.at[index, 'IS_MANAGER'] = 'Ja'
    else:
        newSalesStaff.at[index, 'IS_MANAGER'] = 'Nee'

newSalesStaff.loc[:, ['LAST_NAME', 'POSITION_EN','IS_MANAGER']]

```

Geef met behulp van bovenstaande hulpcode een overzicht met daarin op basis van het aantal jaar dat iemand in dienst is of een medewerker ‘kort in dienst’ (minder dan 25 jaar in dienst) of een ‘lang in dienst’ (groter gelijk dan 12 jaar in dienst) is. Geef daarbij per medewerker in een aparte kolom zowel ‘kort in dienst’ als ‘lang in dienst’ aan. Gebruik (wederom) een for-loop.<br>
(2 kolommen, 102 rijen) 

```python
from datetime import datetime
staffCopy = sales_staff.copy()

for index, row in sales_staff.iterrows():
    date_str = staffCopy.at[index, 'DATE_HIRED']
    date_format = '%d-%m-%Y'
    date_obj = datetime.strptime(date_str, date_format)

    if (date.today().year - date_obj.year) < 25:
        staffCopy.at[index, 'IN_DIENST'] = 'Kort in dienst'
    elif (date.today().year - date_obj.year) >= 25:
        staffCopy.at[index, 'IN_DIENST'] = 'Lang in dienst'

staffCopy.loc[:, ['LAST_NAME', 'DATE_HIRED', 'IN_DIENST']]
```
